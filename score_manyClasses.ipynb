```python

labels = {}
score['高等数学（A）（2-2），公共基础课，必修，学分 5，学时 88，2017-2018-2']=score['高等数学（A）（2-2），公共基础课，必修，学分 5，学时 88，2017-2018-2'].fillna(0).astype(int)
score['高等数学（A）（2-2），公共基础课，必修，学分 5，学时 88，2017-2018-2'] /= 5
score[['高等数学（A）（2-2），公共基础课，必修，学分 5，学时 88，2017-2018-2']] = score[['高等数学（A）（2-2），公共基础课，必修，学分 5，学时 88，2017-2018-2']].astype(int)
t = dict(score['高等数学（A）（2-2），公共基础课，必修，学分 5，学时 88，2017-2018-2'])
for i in t.keys():
    if t[i] not in labels.keys():
        labels[ t[i] ] = 0
    labels[ t[i] ] += 1
```
